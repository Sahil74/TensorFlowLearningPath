```python
#create tensor (3,2) tensor
x = tf.constant([[2,3],[4,5],[7,8]])

y = tf.constant([[21,13],[14,15],[17,18]])
x,y
```

```python
tf.matmul(x,y) #x@y
```

```python
#first matrix 3*2 second 3*2
#they are not matching so we have to reshape
x = tf.reshape(x,shape=(2,3))
```

```python
tf.matmul(x,y) #x@y
```

```python
# can do the transpose #transpose also can change shape
# transpose change the axis == column to raw and raw to column
tf.transpose(x)
```

**The dot product**

you can perform matrix multiplication using

```python
x,y
```

## Aggregating tensors

Aggregating tensors = condensing them from multiple values down to a smaller amount of values.

```python
# get the absolute values = take the negatives and convert into positive numbers
D = tf.constant([-7,-10])
tf.abs(D)
```

##Lets go through the following forms of aggregation
* get the minimum
* max
* mean
* sum

```python
import numpy as np
E = tf.constant(np.random.randint(0,100, size =50))
```

```python
tf.size(E), E.shape, E.ndim
```

```python
tf.reduce_min(E)
tf.reduce_max(E)
tf.reduce_sum(E)
tf.reduce_mean(E)
```

```python
# find the variance of our tensor
import tensorflow_probability as tfp

tfp.stats.variance(E)
tf.math.reduce_variance(tf.cast(E,dtype=tf.float32))
```

```python
# find the std
tf.math.reduce_std(tf.cast(E,dtype= tf.float32))
```

## Find the positional minimum and maximum

```python
# create tensor for Find the positional minimum and maximum
tf.random.set_seed(42)
F = tf.random.uniform(shape = [50])
F
```

```python
# Find the positional maximum #same for min
tf.argmax(F)
```

```python
#index on our largest value position
F[tf.argmax(F)]
```

```python
#check
assert F[tf.argmax(F)] == tf.reduce_max(F)
```
